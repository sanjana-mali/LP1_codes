```python
train=np.array(data.iloc[0:600])
```

```python
train.shape
```

```python
test=np.array(data.iloc[600:])
```

```python
test.shape
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model=GaussianNB()
```

```python
model.fit(train[:,0:8],train[:,8])
```

```python
predicted=model.predict(test[:,0:8])
```

```python
print(test[:,8])
```

```python
print(predicted)
```

```python
count=0
for l in range(168):
    if(predicted[l]==test[l,8]):
        count=count+1
```

```python
print(count)
```

```python
print((count/168)*100)
```
